```python
# Importing Libraries
import pandas as pd
import numpy as np
import sweetviz as sv
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Importing the dataset
default = pd.read_csv('creditcard.csv')
default.head()
```

```python
# Initial EDA- this is not a balanced dataset
report = sv.analyze(default)
report.show_html('sweet_report.html')
```

Highlights from EDA: 
There is no missing values in the dataset. There are also a mix of features (categorical and numerical). I noticed that there are multiple financial variables, such as BILL_AMTs that are skewed right. 


```python
# baseline rd and xgboost models
# Prepare the data for modeling
X = default.drop('default payment next month', axis=1)
y = default['default payment next month']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, shuffle=True, random_state=42)
# Train the random forest model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
# Evaluate the random forest model
y_pred_rf = rf.predict(X_test)
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
# Train the XGBoost model
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)
# Evaluate the XGBoost model
y_pred_xgb = xgb.predict(X_test)
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))

```

```python
# building baseline models with cross validation
rf_cv_scores = cross_val_score(rf, X, y, cv=5)
xgb_cv_scores = cross_val_score(xgb, X, y, cv=5)
print("Random Forest CV Scores:", rf_cv_scores)
print("Random Forest CV Mean Score:", np.mean(rf_cv_scores))
print("XGBoost CV Scores:", xgb_cv_scores)
print("XGBoost CV Mean Score:", np.mean(xgb_cv_scores))

#printing the standard deviations
print("Random Forest CV Mean:", np.mean(rf_cv_scores))
print("Random Forest CV Std:", np.std(rf_cv_scores))

print("XGBoost CV Mean:", np.mean(xgb_cv_scores))
print("XGBoost CV Std:", np.std(xgb_cv_scores))
```

Both RF and XGBoost show very similar performance so far, but RF slightly outperforms XGBoost in terms of accuracy. The standard deviations are also very low (about 0.003) for both models. The different between the two models' performances are minimal so far. The mean score represents the average performance across the five folds, and the standard deviation indicates how consistent the model is across different subsets of data. 

```python
# repeat baseline model training but use recall as teh evaluation metrics
rf_cv_recall_scores = cross_val_score(rf, X, y, cv=5, scoring='recall')
xgb_cv_recall_scores = cross_val_score(xgb, X, y, cv=5, scoring='recall')
print("Random Forest CV Recall Scores:", rf_cv_recall_scores)
print("Random Forest CV Mean Recall Score:", np.mean(rf_cv_recall_scores))
print("XGBoost CV Recall Scores:", xgb_cv_recall_scores)
print("XGBoost CV Mean Recall Score:", np.mean(xgb_cv_recall_scores))   
print(classification_report(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_xgb))
```

Although both models seem to perform pretty well with accuracy, I see that recall is only about 38%. Looking at accuracy alone would be misleading when evaluating the model performance. This may be due to the class imbalance that we saw from the initial EDA report. 

```python
# rebuild rf and xgboost models with adjustment to class weights to address class imbalance
rf_weighted = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_weighted.fit(X_train, y_train)
y_pred_rf_weighted = rf_weighted.predict(X_test)
print("Random Forest with Class Weights Classification Report:")
print(classification_report(y_test, y_pred_rf_weighted))
xgb_weighted = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', scale_pos_weight= (y_train == 0).sum() / (y_train == 1).sum())

xgb_weighted.fit(X_train, y_train)
y_pred_xgb_weighted = xgb_weighted.predict(X_test)
print("XGBoost with Class Weights Classification Report:")
print(classification_report(y_test, y_pred_xgb_weighted))
```

RF model performance got slightly worse when it comes to recall (0.38 --> 0.36), but saw no major difference. XGBoost saw a big improvement (0.38 --> 0.58) in recall. However, accuracy and precision did go down. This shows that there are tradeoffs. 

```python
# hyperparameter tuning with grid search for random forest with 2 hyperparameters and 2 values for each
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10]
}
grid_search_rf = GridSearchCV(estimator=rf_weighted, param_grid=param_grid_rf, cv=5, scoring='recall')
grid_search_rf.fit(X_train, y_train)
print("Best Hyperparameters for Random Forest:", grid_search_rf.best_params_)
best_rf = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)
print("Best Random Forest Classification Report:")
print(classification_report(y_test, y_pred_best_rf))    
```

```python
# hyperparameter tuning with grid search for xgboost with 2 hyperparameters and 2 values for each
from pytest import param


param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6]
}
grid_search_xgb = GridSearchCV(estimator=xgb_weighted, param_grid=param_grid_xgb, cv=5, scoring='recall')
grid_search_xgb.fit(X_train, y_train)
print("Best Hyperparameters for XGBoost:", grid_search_xgb.best_params_)
best_xgb = grid_search_xgb.best_estimator_
y_pred_best_xgb = best_xgb.predict(X_test)
print("Best XGBoost Classification Report:") 
print(classification_report(y_test, y_pred_best_xgb))     
```

After tuning, the models seemed to both improve. For RF, recall improved from 0.36 to 0.56. F1 improved from 0.46 to 0.53. For XG, recall improved from 0.58 to 0.64 and F1 improved from 0.52 to 0.54. Tuning the XG boost did make it improve, but not as much of a difference as seen in the RF model.
